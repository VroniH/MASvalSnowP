```python
def create_pit_values(pit):
    #pit = Pit1
    normedarray = np.arange(0,1+0.005,0.005)[::-1]
    values = np.zeros(len(normedarray))
    for df_index, entry in enumerate(pit["normedTop"]):
        for index, value in enumerate(normedarray):
            if entry>=value:
                values[index] = pit["haerte"].values[df_index]
    return np.transpose([values])
            
```

```python
file_name = "pits_harndess_normed"

pits = [Pit2, Pit6, Pit7, Pit9, Pit1, Pit3, Pit4, Pit5, Pit8, Pit10, Pit11]
xticklabels = [2,6,7,9,1,3,4,5,8,10,11]


plt.style.use('ggplot')
fig, axes = plt.subplots(1, len(pits), figsize=(10, 5))
fig.subplots_adjust(left=.08, bottom=.08, right=.97, top=.92)


for index, pit in enumerate(pits):
    values = create_pit_values(pit)
    hoehe = Pit1["vonhoehe"].values[0]
    #cax = axes[index].imshow(values, extent=[0,.1,0,hoehe], vmin=1, vmax=6, cmap=plt.get_cmap("Blues",11))
    cax = axes[index].imshow(values, extent=[0,.1,0,1], vmin=1, vmax=6, cmap=plt.get_cmap("Blues",11))
    axes[index].grid(False)
    axes[index].set_title("Pit %d" % (xticklabels[index]), y=-0.07)
    axes[index].set_xticks([])
    axes[index].set_yticks([])

    
axes[0].set_ylabel("Normalized height")
axes[0].set_yticks(np.linspace(0,1,11, endpoint=True))


hardnessnames = ["F","F - 4F", "4F", "4F - 1F", "1F", "1F - P", "P", "P - K", "K", "K - I", "I"]
cbar = fig.colorbar(cax, ax=axes.ravel().tolist(), ticks=np.linspace(1,6,12,endpoint=True) + 0.24, orientation='vertical')
cbar.set_label("Hardness")
cbar.ax.set_yticklabels(hardnessnames)

xpos = 0.343
line = Line2D((xpos,xpos),(0,1), transform=fig.transFigure, color="r")
fig.lines =  line,


# fig.savefg("plots/%s.eps" % file_name)
# fig.savefig("plots/%s.png" % file_name, dpi=200)
```

```python

plt.style.use('ggplot')

file_name = "pits_harndess"
norm = matplotlib.colors.Normalize(vmin=1, vmax=6, clip=True)
mapper = cm.ScalarMappable(norm=norm, cmap=plt.get_cmap("Blues",11))

pits = [Pit2, Pit6, Pit7, Pit9, Pit1, Pit3, Pit4, Pit5, Pit8, Pit10, Pit11]

fig, axes = plt.subplots(1, 1, figsize=(12, 6))
fig.subplots_adjust(left=.1, bottom=.05, right=.98, top=.98)

for index, pit in enumerate(pits):
    for von, bis, haerte in zip(pit["vonhoehe"],pit["bishoehe"],pit["haerte"]):
        color = mapper.to_rgba(haerte)
        plt.bar(x=index+.6 ,bottom=bis, height=von-bis, left=0.0, width=0.8, color=color)
        
hardnessnames = ["F","F - 4F", "4F", "4F - 1F", "1F", "1F - P", "P", "P - K", "K", "K - I", "I"]
cbar = fig.colorbar(cax, ax=axes, ticks=np.linspace(1,6,12,endpoint=True) + 0.21, orientation='vertical')
#cbar.set_label("Hardness")
cbar.ax.set_yticklabels(hardnessnames)

axes.set_ylabel("Snow height $(\\mathrm{cm})$")
ystep = 20
yticks = np.arange(0, 310+ystep, ystep)
axes.set_yticks(yticks)

xticklabels = ["Pit %d" % i for i in [2,6,7,9,1,3,4,5,8,10,11]]
axes.set_xticks(np.linspace(0,10,11,endpoint=True)+1.0)
axes.set_xticklabels(xticklabels)
axes.xaxis.grid(False)
axes.yaxis.grid(True, color=".5", linestyle="-")

axes.set_facecolor((1.0, 1.0, 1.0))
axes.tick_params(axis='both', top=False, right=False, left=False, bottom=True)
axes.spines['bottom'].set_color('0.0')

xpos = 0.364
line = Line2D((xpos,xpos),(0,0.94), transform=fig.transFigure, color="r")
fig.lines =  line,

plt.show()
fig.savefig("plots/%s.eps" % file_name)
fig.savefig("plots/%s.png" % file_name, dpi=200)
```
